```python
def saudacao(nome):
  print(f"Olá, {nome}!")

nome = input("Digite seu nome: ")
saudacao(nome)
```

```python
def calcularDobro(numero):
  return numero * 2

numero = int(input("Digite um número: "))
print(f"O dobro de {numero} é {calcularDobro(numero)}")
```

```python
def ePar(numero):
  if numero % 2 == 0:
    return True
  else:
    return False

numero = int(input("Digite um número: "))
if ePar(numero):
  print(f"{numero} é par")
else:
  print(f"{numero} é ímpar")
```

```python
def calcularArea(base, altura):
  return base * altura

base = float(input("Digite a base do retângulo: "))
altura = float(input("Digite a altura do retângulo: "))
print(f"A área do retângulo é {calcularArea(base, altura)}")
```

```python
def gerarEmail(nome):
  return (f"{nome.lower()}@senai.com").replace(" ", ".").lower()

nome = input("Digite seu nome: ")
print(f"Seu email é {gerarEmail(nome)}")
```

```python
def calculoPreco(preco, desconto):
  return preco * ((100 - desconto) / 100)

preco = float(input("Digite o preço: R$"))
desconto = int(input("Digite a porcentagem do desconto: "))

print(f"Novo preço: R${calculoPreco(preco, desconto):.2f}")
```

```python
def validarSenha(senha):
  if len(senha) >= 8:
    print(f"Senha válida")
  else:
    print(f"Senha inválida")

senha = input(f"Digite a senha: ")
validarSenha(senha)
```

```python
def analisarClima(temperatura):
  if temperatura >= 25:
    print(f"Está quente 🔥 ({temperatura:.1f}°C)")
  elif 15 < temperatura < 25:
    print(f"A temperatura está agradável ({temperatura:.1f}°C)")
  else:
    print(f"Está frio ❄️ ({temperatura:.1f}°C)")

temperatura = float(input(f"Qual a temperatura atual? "))
analisarClima(temperatura)
```

```python
def avaliarEmprestimo(salario, parcela):
  if parcela <= salario * 0.3:
    print(f"Empréstimo aprovado")
  else:
    print(f"Empréstimo negado")

salario = float(input(f"Digite o salário do funcionário: R$"))
parcela = float(input(f"Digite o valor da parcela: R$"))
avaliarEmprestimo(salario, parcela)
```

```python
def IMC(peso, altura):
  imc = peso / altura ** 2
  print(f"Seu IMC é {imc:.2f}")
  if imc < 18.5:
    print(f"Você está abaixo do peso")
  elif 18.5 <= imc < 25:
    print(f"Você está no peso ideal")
  elif 25 <= imc < 30:
    print(f"Você está com sobrepeso")
  elif 30 <= imc < 35:
    print(f"Você está com obesidade grau I")
  elif 35 <= imc < 40:
    print(f"Você está com obesidade grau II")
  else:
    print(f"Você está com obesidade grau III")

peso = float(input(f"Digite seu peso: "))
altura = float(input(f"Digite sua altura: "))
IMC(peso, altura)
```

```python
def calculadora():
  print("--- Calculadora ---")
  opcao = input("1. Adição\n2. Subtração\n3. Multiplicação\n4. Divisão\n5. Exponenciação\n6. Radiciação\n7. Sair\nEscolha uma opção: ").lower()

  if opcao in ["7", "sair"]:
    print("Saindo da calculadora. Até mais!")
    return False

  if opcao in ["1", "adicao", "2", "subtracao", "3", "multiplicacao", "4", "divisao", "5", "exponenciacao", "6", "radiciacao"]:
    n1 = float(input("Digite o primeiro número: "))
    n2 = float(input("Digite o segundo número: "))
  else:
    print("Operação desconhecida. Tente novamente.")
    return True;

  match opcao:
    case "1" | "adicao":
      print(f"{n1} + {n2} = {n1 + n2}")
    case "2" | "subtracao":
      print(f"{n1} - {n2} = {n1 - n2}")
    case "3" | "multiplicacao":
      print(f"{n1} * {n2} = {n1 * n2}")
    case "4" | "divisao":
      if n2 == 0:
        print("Erro: Não é possível dividir por zero!")
      else:
        print(f"{n1} / {n2} = {n1 / n2}")
    case "5" | "exponenciacao":
      print(f"{n1} ** {n2} = {n1 ** n2}")
    case "6" | "radiciacao":
      if n2 == 0:
        print("Erro: O índice da raiz não pode ser zero!")
      else:
        print(f"{n2:.0f}√{n1:.0f} = {(n1 ** (1 / n2)):.4f}")

  return True

continuar = True
while continuar:
    continuar = calculadora()
```

```python
saldo = 1000

def realizarSaque(saldo, saque):
  if saque > saldo:
    print("Saldo insuficiente, tente em outro momento...")
  else:
    saldo -= saque
    print(f"Você sacou R${saque:.2f} com sucesso!")
  return saldo

continuar = True
while continuar:
  opcao = input(f"Deseja ir no banco?\nSaldo atual: R${saldo:.2f}\n1 - Sim\n2 - Não\nOpção: ").lower()
  if opcao in ["sim", "s", "yes", "y", "1"] and saldo > 0:
    saque = float(input("Quanto deseja sacar? R$"))
    saldo = realizarSaque(saldo, saque)
  else:
    print("Indo embora...")
    continuar = False
```

```python
def converterMoeda(valor, cotacao):
  return valor * cotacao

continuar = True
while continuar:
  opcao = input(f"Qual moeda deseja converter?\n1 - Dólar\n2 - Euro\n3 - Sair\nOpção: ").lower()
  if opcao == "3" or opcao == "sair":
    print(f"Saindo...")
    continuar = False
  else:
    valor = float(input(f"Por qual valor? R$"))
    match opcao:
      case "1" | "dolar":
        cotacao = 5.00
      case "2" | "euro":
        cotacao = 6.00
      case _:
        cotacao = 1.00
    print(f"R${valor:.2f} = ${converterMoeda(valor, cotacao):.2f}")
```

```python
def criptografarTexto(texto):
  novaPalavra = ""
  vogais = "aeiouAEIOU"

  for letra in texto:
    if letra in vogais:
      novaPalavra += "*"
    else:
      novaPalavra += letra
  return novaPalavra.replace(" ", "#")

texto = input("Digite um texto: ")
print(f"Texto criptografado: {criptografarTexto(texto)}")
```
