```python
for i in range(1, 51):
  if i % 3 == 0 and i % 5 == 0:
    print(f"FizzBuzz!")
  elif i % 3 == 0:
    print(f"Fizz")
  elif i % 5 == 0:
    print(f"Buzz")
  else:
    print(f"{i}")
```

```python
caracol = 0
dia = 0

while caracol < 10:
  caracol += 3
  print(f"O caracol subiu 3 metros durante o dia ({caracol}m)")
  if caracol >= 10:
    print(f"E assim... no {dia}° dia ele conseguiu sair do poço!")
    break;
  caracol -= 2
  print(f"Mas caiu 2 metros durante a noite ({caracol}m)\nNo próximo dia...")
  dia += 1
```

```python
bolinha = 20
quicada = 0
print(f"Altura inicial da bolinha: {bolinha}m")

while bolinha > 0.1:
  bolinha /= 2
  quicada += 1
  print(f"{quicada + 1}° Altura atingida: {bolinha:.1f}m")

print(f"A bolinha quicou {quicada} vezes")
```

```python
def ePrimo(num):
  if num <= 1:
    return False
  for i in range(2, int(num ** 0.5) + 1):
    if num % i == 0:
      return False
  return True

num = int(input("Escolha um número inteiro: "))
print(f"Seu número é primo? {ePrimo(num)}")
```

```python
def calcularNotas(valor):
  print(f"--- Contagem de notas ---\nNotas de 50: {valor // 50}")
  valor = valor % 50
  print(f"Notas de 20: {valor // 20}")
  valor = valor % 20
  print(f"Notas de 10: {valor // 10}")
  valor = valor % 10
  print(f"Notas de 1: {valor // 1}")

valor = int(input(f"Digite o valor: R$"))
calcularNotas(valor)
```

```python
palavra = input("Digite uma palavra: ").lower()
palavraInvertida = ""

for i in range(len(palavra) - 1, -1, -1):
  palavraInvertida += palavra[i]

print(f"A palavra invertida é: {palavraInvertida}")

if palavra == palavraInvertida:
  print("A palavra é um palíndromo")
else:
  print("A palavra não é um palíndromo")
```

```python
salario = 1000

def calcularSalario(salario):
  # 1º Mês: Aplica o bônus de 10%
  # R$ 1.000,00 + (10% de R$ 1.000,00) -> 1.000,00 + 100,00 = R$ 1.100,00
  salario_mes1 = salario * 1.10

  # 2º Mês: Aplica o desconto de 10% sobre o salário atual (R$ 1.100,00)
  # R$ 1.100,00 - (10% de R$ 1.100,00) -> 1.100,00 - 110,00 = R$ 990,00
  salario_mes2 = salario_mes1 * 0.90
  return salario_mes2

print(f"Salário inicial: R${salario:.2f}")
print(f"Salário final: R${calcularSalario(salario):.2f}")
```

```python
def eBissexto(ano):
  if ano % 4 == 0 and (ano % 100 != 0 or ano % 400 == 0):
    print(f"{ano} Bissexto!")
  else:
    print(ano)

inicio = int(input("Digite um início: "))
fim = int(input("Digite um fim: "))

for i in range(inicio, fim + 1):
  eBissexto(i)
```

```python
num = int(input("Digite um número inteiro: "))

num_original = num
soma = 0

while num > 0:
    digito = num % 10
    soma += digito
    num = num // 10

num = num_original
i = 1

while num > 0:
  if num >= 10:
    digito = num // (10 ** (len(str(num)) - 1))
    num = num % (10 ** (len(str(num)) - 1))
    print(f"{i}° - {digito}")
    i += 1
  else:
    digito = num
    print(f"{i}° - {digito}")
    break;

print(f"A soma dos dígitos de {num_original} é: {soma}")
```
