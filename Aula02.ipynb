```python
saldo = 500

while True:
  opcao = input(f"Deseja ir no banco?\nSaldo atual: R${saldo:.2f}\n1 - Sim\n2 - Não\nOpção: ").lower()
  if opcao in ["sim", "s", "yes", "y", "1"] and saldo > 0:
    valor = float(input("Quanto deseja sacar? R$"))
    if valor >= saldo:
      print("Saldo insuficiente, tente novamente...")
    else:
      saldo -= valor
      print(f"Você sacou R${valor:.2f} com sucesso!")
  else:
    print("Indo embora...")
    break;
```

```python
total = [0] * 11

while True:
    nota = int(input("Qual nota você dá para esse programa? (0 para sair): "))
    if nota == 0:
        print("Saindo do sistema...\n")
        break;
    elif nota < 1 or nota > 10:
        print(f"Nota inválida ({nota}), tente novamente...")
    else:
        total[nota] += 1
        print(f"Nota {nota} adicionada com sucesso!")

print("--- Notas finais ---")
for i in range(1, 11):
  if total[i] == 0:
    print(f"Nota {i} = Nenhum voto")
  else:
    print(f"Nota {i} = {total[i]} voto(s)")
```
