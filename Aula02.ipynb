```python
numero = int(input("Digite um número inteiro: "))

if numero >= 0:
  print(f"O número {numero} é maior ou igual a zero (positivo)")
else:
  print(f"O número {numero} é menor que zero (negativo)")
```

```python
lado1 = float(input("Valor do primeiro lado: "))
lado2 = float(input("Valor do primeiro lado: "))
lado3 = float(input("Valor do primeiro lado: "))

if lado1 + lado2 >= lado3 or lado1 + lado3 >= lado2 or lado2 + lado3 >= lado1:
  if lado1 == lado2 == lado3:
    print(f"Ele é um triângulo equilátero")
  elif lado1 == lado2 != lado3 or lado1 != lado2 == lado3 or lado1 == lado3 != lado2:
    print(f"Ele é um triângulo isósceles")
  else:
    print(f"Ele é um triângulo escaleno")
else:
  print(f"Não forma um triângulo")
```

```python
numFinal = int(input("Número final da placa: "))

match numFinal:
  case 1 | 2:
    print(f"O carro não roda de Segunda")
  case 3 | 4:
    print(f"O carro não roda de Terça")
  case 5 | 6:
    print(f"O carro não roda de Quarta")
  case 7 | 8:
    print(f"O carro não roda de Quinta")
  case 9 | 0:
    print(f"O carro não roda de Sexta")
  case _:
    print(f"Outro dia ou número inválido")
```

```python
mes = int(input("Digite o número de um mês: "))

match mes:
  case 1 | 2 | 3:
    print(f"O {mes}° mês do ano está presente no 1° Trimestre")
  case 4 | 5 | 6:
    print(f"O {mes}° mês do ano está presente no 2° Trimestre")
  case 7 | 8 | 9:
    print(f"O {mes}° mês do ano está presente no 3° Trimestre")
  case 10 | 11 | 12:
    print(f"O {mes}° mês do ano está presente no 4° Trimestre")
  case _:
    print(f"Mês inválido")
```

#**Laços de Repetição**

```python
#for
inicio = int(input("Defina um início: "))
fim = int(input("Defina um fim: "))

for i in range(inicio, fim + 1):
  print(f"{i}")
```

```python
#while
inicio = int(input("Defina um início: "))
fim = int(input("Defina um fim: "))

while inicio < fim + 1:
  print(f"{inicio}")
  inicio+=1
```
